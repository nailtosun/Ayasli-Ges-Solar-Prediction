# Pulling Data using DarkSky API

Required librarires

```python
import requests
import pandas as pd
import datetime
```

```python
i = 0 #initialize counter
#final-time 1533115500
#stopped-time 1525376400
#starting-time 1521061200
loop = (1525376400-1521061200)/300
print(loop)
```

## First data

Pulling the first data is useful for checking the right format before large amount pulling.

```python
initial_time = 1525376400 #this is the starting time in UNIX format
timestamp = initial_time + 300*i #next request is after 5 minutes
url = 'https://api.darksky.net/forecast/6829167db76975d957b2149428e13801/39.8906,32.7817,' 
url = url + str(timestamp)
data = requests.get(url) #data returned in json format
data=data.json() #opening json 

#Sometimes there may be some missing values from Darksky. Try-except blocks is useful for stopping issues about KeyError

try:
    cloudCover= data['currently']['cloudCover']
except KeyError:
    cloudCover = None  
try:
    temp= data['currently']['apparentTemperature']
except KeyError:
    temp = None
try:
    humidity = data['currently']['humidity']
except KeyError:
    humidity = None
try:
    icon = data['currently']['icon']
except KeyError:
    icon = None
try:
    windSpeed = data['currently']['windSpeed']
except KeyError:
    windSpeed = None
try:
    visibility = data['currently']['visibility']
except KeyError:
    visibility = None
try:
    windBearing = data['currently']['windBearing']
except KeyError:
    windBearing = None
try:
    pressure = data['currently']['pressure']
except KeyError:
    pressure = None
try:
    dewPoint = data['currently']['dewPoint']
except KeyError:
    dewPoint = None

row = {'cloudCover':cloudCover , 'temp' : temp , 'humidity' : humidity , 'icon' : icon,'windSpeed' : windSpeed , 'visibility' : visibility, 'windBearing' : windBearing ,'pressure' : pressure ,"dewPoint" : dewPoint}
time_index = datetime.datetime.fromtimestamp(int(timestamp)).strftime('%Y-%m-%d %H:%M:%S')
df2 = pd.DataFrame(data=row,index=[timestamp])
```

## For loop

```python
for i in range (0,int(loop)): #one
    initial_time = 1525376400
    timestamp = initial_time + 300*i #5 minute
    url = 'https://api.darksky.net/forecast/18971c55f948a8d2261f41457e4db0cb/39.8906,32.7817,' 
    url = url + str(timestamp)
    data = requests.get(url)
    data=data.json()
    try:
        cloudCover= data['currently']['cloudCover']
    except KeyError:
        cloudCover = None  
    try:
        temp= data['currently']['apparentTemperature']
    except KeyError:
        temp = None
    try:
        humidity = data['currently']['humidity']
    except KeyError:
        humidity = None
    try:
        icon = data['currently']['icon']
    except KeyError:
        icon = None
    try:
        windSpeed = data['currently']['windSpeed']
    except KeyError:
        windSpeed = None
    try:
        visibility = data['currently']['visibility']
    except KeyError:
        visibility = None
    try:
        windBearing = data['currently']['windBearing']
    except KeyError:
        windBearing = None
    try:
        pressure = data['currently']['pressure']
    except KeyError:
        pressure = None
    try:
        dewPoint = data['currently']['dewPoint']
    except KeyError:
        dewPoint = None

    row = {'cloudCover':cloudCover , 'temp' : temp , 'humidity' : humidity , 'icon' : icon,'windSpeed' : windSpeed , 'visibility' : visibility, 'windBearing' : windBearing ,'pressure' : pressure ,"dewPoint" : dewPoint}
    time_index = datetime.datetime.fromtimestamp(int(timestamp)).strftime('%Y-%m-%d %H:%M:%S')
    df = pd.DataFrame(data=row,index=[timestamp])
    df2 = df2.append(df)
df2.to_csv('sonpull2.csv')
```
